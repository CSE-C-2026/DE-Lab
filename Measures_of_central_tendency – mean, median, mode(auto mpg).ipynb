```python
# To obtain only the numeric columns
numeric_columns = data.select_dtypes(include="number")
print("Numeric columns:", numeric_columns)
```

```python
#To print the quartiles
print(data.quantile([0.25,0.5,0.75],numeric_only=True))
```

```python
#to find the IQR for the data
print(data.quantile(0.75,numeric_only=True) - data.quantile(0.25,numeric_only=True))
```

```python
# TO convert data into dataframec
numeric_data=pd.DataFrame(numeric_columns)
numeric_data
```

```python
# Mean
for i in numeric_data:
  d=data[i]
  mean=sum(d) / len(d)
  print(f"{i} : {mean}")
```

```python
# Median
for i in numeric_data:
  d=data[i].tolist()
  d.sort()
  n=len(d)
  if n%2==0:
    median=(d[n//2 - 1] + d[n//2]) / 2
  else:
    median=d[n//2]
  print(f"{i} : {median}")
```

```python
#Mode
from collections import Counter
for i in numeric_data:
  d=data[i]
  freq=Counter(d)
  max_val=max(freq.values())
  mode=[k for k,v in freq.items() if v==max_val]
  print(f"{i} : {mode}")
```

```python
#Variance
for i in numeric_data:
  d=data[i]
  mean=sum(d)/len(d)
  var=sum((x-mean)**2 for x in d) / len(d)
  print(f"{i} : {var}")
```

```python
#Standard Deviation
for i in numeric_data:
  d=data[i]
  mean=sum(d)/len(d)
  var=sum((x-mean)**2 for x in d) / len(d)
  std=var**0.5
  print(f"{i} : {std}")
```
