```python
import pandas as pd
data=pd.read_csv('iris.csv')
```

```python
#Box plot for each predictor
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.boxplot(data)
plt.title("Box Plot for Each Predictor in Iris Dataset")
plt.show()
```

```python
#box plot for a specific feature
plt.figure(figsize=(6, 4))
sns.boxplot(y=data['petal.length'])
plt.title("Box Plot for Petal Length")
plt.show()
```

```python
#Histogram for sepal width
plt.figure(figsize=(6, 4))
sns.histplot(data['sepal.width'], bins=15, kde=True)  # kde=True adds a density curve
plt.title("Histogram of Sepal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
```

```python
#Scatter plot of petal length vs. sepal length
plt.figure(figsize=(6, 4))
sns.scatterplot(x=data['sepal.length'], y=data['petal.length'])
plt.title("Scatter Plot of Petal Length vs. Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()
plt.show()
```
