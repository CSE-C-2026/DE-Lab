```python
# 5. Change columns of a DataFrame
# Renaming column 'Pregnancies' to 'NumPregnancies' as an example
data.rename(columns={'Pregnancies': 'NumPregnancies'}, inplace=True)
print("Updated column names:", data.columns.tolist())
```

```python
# 7. Bind sets of rows of DataFrames
# Creating another DataFrame for demonstration
data_extra = data.sample(5)
combined_rows = pd.concat([data, data_extra], axis=0)
print("Dataset after binding rows:\n", combined_rows)
```

```python
# 8. Bind sets of columns of DataFrames
# Creating another DataFrame with additional columns for demonstration
extra_columns = pd.DataFrame({
    'NewColumn1': range(data.shape[0]),
    'NewColumn2': range(data.shape[0], 2 * data.shape[0])
})
combined_columns = pd.concat([data, extra_columns], axis=1)
print("Dataset after binding columns:\n", combined_columns)
```
